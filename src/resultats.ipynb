```python
import pandas as pd
import numpy as np  
import matplotlib.pyplot as plt
```

```python
# Load data 
data = np.loadtxt('Scores_titFortat_stochastique.csv',delimiter=';',dtype=float)
```

```python
# print(len(data[data[:,0] < data[:,1]]))
# print(len(data[data[:,0] > data[:,1]]))
# print(len(data[data[:,0] == data[:,1]]))
data[:]
```

```python
# Create a figure with matplotlib 12*15"
fig = plt.figure(figsize=(15, 12)) 

# Create a barplot with the sum of each column
plt.bar(["TitForTat","Stochastique"], [np.sum(data[:,0]),np.sum(data[:,1])], color=['#0066ff','#ff0000'])
# Title : "Points marqués par chaque équipe"
plt.title("Parties gagnées par chaque stratégie")
# First bar legend : "Points marqués par l'équipe 1"
plt.annotate(np.sum(data[:,0]), xy=(0, np.sum(data[:,0])), xytext=(0, 0), textcoords='offset points', ha='center', va='bottom', fontsize=12)
# Second bar legend : "Points marqués par l'équipe 2"
plt.annotate(np.sum(data[:,1]), xy=(1, np.sum(data[:,1])), xytext=(0, 0), textcoords='offset points', ha='center', va='bottom', fontsize=12)
# Save the figure
plt.savefig('Scores_titFortat_stochastique.png')
```
